```python
import librosa
import librosa.display
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
import tensorflow as tf
```

```python
train_file = pd.read_csv('train/train.csv')
print(train_file.shape)
print(train_file.head())
print(train_file.Class.unique())
```

```python
i = 1
mfccs = []
labels = []
audios  =[]
for index, label in zip(train_file.ID, train_file.Class):
    if(i==2000):
        break
    print(i, label , index)
    i = i+1
    path = 'train/Train/'+str(index)+'.wav'
    audio, sr = librosa.load(path)
#     print(audio.shape)
    audios.append(audio)
    labels.append(label)
    
aud = tf.keras.preprocessing.sequence.pad_sequences(audios,maxlen= 88200,dtype = 'float32')
print(aud.shape) 

for audio, label in zip(aud, labels):
    spec = librosa.power_to_db(librosa.feature.melspectrogram(audio,   
                                                              sr = sr ,n_fft=2000, 
                                                              hop_length=512+500,
                                                              n_mels=88))### This is new
    mfccs.append(spec)
    ### This I used earlier it doesn';t work good in cv set
#     stft = np.abs(librosa.stft(audio, n_fft= 3000 , win_length = 2775))
#     spec = librosa.feature.melspectrogram(S = stft)
#     mfcc = librosa.feature.mfcc(S = spec, n_mfcc= 128)
# #     print(mfcc.shape)
#     mfccs.append(mfcc)
```

```python
mfccs = np.array(mfccs)
labels = np.array(labels)
# mfccs = np.reshape()
print(mfccs.shape , labels.shape)
```

```python
# labels[0:10]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(labels)
from tensorflow.keras.utils import to_categorical
y_cat = to_categorical(y , 10)
print(y_cat.shape)
```

```python
X = np.reshape(mfccs , (mfccs.shape[0] , mfccs.shape[1], mfccs.shape[2], 1 ))
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
x_tr , x_cv , y_tr , y_cv = train_test_split( X , y_cat , test_size = 0.2 )
print(x_tr.shape , x_cv.shape , y_tr.shape ,y_cv.shape )
```

```python
from tensorflow.keras.layers import Conv2D , Dense , Flatten , MaxPooling2D, Dropout, BatchNormalization, Input
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import Sequential, Model
```

```python
model = Sequential()


model.add(Conv2D(filters=32,kernel_size=(3,3),padding= 'same',
                 input_shape=(mfccs.shape[1],mfccs.shape[2],1),activation='relu'))
model.add(Conv2D(filters=32,kernel_size=(3,3),padding= 'same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization(axis=-1))

model.add(Conv2D(filters=64,kernel_size=(3,3),padding= 'same',activation='relu'))
model.add(Conv2D(filters=64,kernel_size=(3,3),padding= 'same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization(axis=-1))

model.add(Conv2D(filters=128,kernel_size=(3,3),padding= 'same',activation='relu'))
model.add(Conv2D(filters=128,kernel_size=(3,3),padding= 'same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=256,kernel_size=(3,3),padding= 'same',activation='relu'))
model.add(Conv2D(filters=256,kernel_size=(3,3),padding= 'same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
# model.add(Dense(16384 , activation='relu'))
model.add(Dense(256 , activation='relu'))
model.add(Dense(10, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.0001),
             loss=tf.keras.losses.categorical_crossentropy,
             metrics=['accuracy'])
callbacks = [
    EarlyStopping(patience=2 , monitor='val_acc')
]
```

```python
model.fit(x_tr , y_tr, epochs=10, callbacks= callbacks ,batch_size=100, validation_data=(x_cv , y_cv))
```
